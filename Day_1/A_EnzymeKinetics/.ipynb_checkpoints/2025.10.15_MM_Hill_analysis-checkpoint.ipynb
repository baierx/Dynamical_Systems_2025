# Nonlinear Kinetics: Michaelis-Menten & Hill function

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Simple Implementation

```python
def my_rate(S, n):

    return (1 * S**n) / (1**n + S**n)
    
dim_1, dim_2 = 10, 12

S_values = np.linspace(0, 3, dim_1)  # substrate from 0 to 4
n_values = np.linspace(0.5, 4, dim_2)  # n from 0.5 to 4

# Create meshgrid for 3D plotting
S, N = np.meshgrid(S_values, n_values)

rate_landscape = my_rate(S, N)

fig, ax = plt.subplots(figsize=(4, 4))

ax.imshow(rate_landscape, aspect='auto');

rate_landscape.shape
```

## Fulll Implementation

```python
def hill_kinetics(S, k_max, K_m, n):
    """
    Hill kinetics equation
    S: substrate concentration
    k_max: maximum reaction rate
    K_m: Michaelis constant (fixed at 1)
    n: Hill coefficient (cooperativity exponent)
    """
    return (k_max * S**n) / (K_m**n + S**n)
```

```python
# Set fixed parameters
K_m = 1
k_max_fixed = 1

# Create ranges for substrate and Hill coefficient
S_values = np.linspace(0, 3, 100)  # substrate from 0 to 4
n_values = np.linspace(0.5, 4, 100)  # n from 0.5 to 4

# Create meshgrid for 3D plotting
S, N = np.meshgrid(S_values, n_values)

# Calculate reaction rate landscape
rate_landscape = np.zeros_like(S)

for index1 in range(S.shape[0]):
    for index2 in range(S.shape[1]):
        rate_landscape[index1, index2] = hill_kinetics(S[index1, index2], k_max_fixed, K_m, N[index1, index2])

rate_landscape_gradient = np.gradient(rate_landscape)

# Create the visualization
fig = plt.figure(figsize=(15, 10))

# 1. Heatmap of rate landscape (S vs n)
ax1 = fig.add_subplot(2, 2, 1)

im1 = ax1.imshow(rate_landscape, extent=[S_values.min(), S_values.max(), 
                                       n_values.min(), n_values.max()],
               origin='lower', aspect='auto', cmap='viridis')
ax1.set_xlabel('Substrate Concentration (S)')
ax1.set_ylabel('Hill Coefficient (n)')
ax1.set_title(f'Rate Heatmap\n(S vs n, k_max={k_max_fixed})')
plt.colorbar(im1, ax=ax1, label='Reaction Rate', shrink=0.5)

# Add contour lines to heatmap
contour_levels = [0.1, 0.3, 0.5, 0.7, 0.9]
contours = ax1.contour(S, N, rate_landscape, levels=contour_levels, 
                       colors='white', alpha=0.7, linewidths=1)
ax1.clabel(contours, inline=True, fontsize=8, fmt='%.1f')


# 2. 3D Landscape plot - Rate vs S and n
# ax2 = fig.add_subplot(2, 2, 2, projection='3d')
ax2 = fig.add_subplot(2, 2, 2)

im2 = ax2.imshow(rate_landscape_gradient[1], extent=[S_values.min(), S_values.max(), 
                                       n_values.min(), n_values.max()], vmax=0.03,
               origin='lower', aspect='auto', cmap='viridis')
ax2.set_xlabel('Substrate Concentration (S)')
ax2.set_ylabel('Hill Coefficient (n)')
ax2.set_title(f'Change of Rate Landscape\n(S vs n, k_max={k_max_fixed}, K_m={K_m})')
plt.colorbar(im2, ax=ax2, label='Reaction Rate', shrink=0.5)

# Add contour lines to heatmap
contour_levels_gradient = [0.005, 0.01, 0.02, 0.03]
contours_grad = ax2.contour(S, N, rate_landscape_gradient[1], levels=contour_levels_gradient, 
                       colors='white', alpha=0.7, linewidths=1)
ax2.clabel(contours_grad, inline=True, fontsize=8, fmt='%.3f')



# 3. Cooperativity
ax3 = fig.add_subplot(2, 2, 3)

# Show how the curve shape changes with n
S_fine = np.linspace(0, 3.5, 300)
n_transition = [0.5, 1.0, 2.0, 4.0]
colors_transition = ['blue', 'green', 'orange', 'red']

for n, color in zip(n_transition, colors_transition):
    rate = hill_kinetics(S_fine, k_max_fixed, K_m, n)
    ax3.plot(S_fine, rate, color=color, linewidth=3, label=f'n={n}')

# Add region annotations
ax3.fill_between(S_fine, 0, 1, where=(S_fine<1), alpha=0.1, color='blue', label='S < K_m')
ax3.fill_between(S_fine, 0, 1, where=(S_fine>=1), alpha=0.1, color='red', label='S  K_m')
ax3.axvline(x=1, color='black', linestyle='--', alpha=0.5, label='S=K_m=1')

ax3.set_xlabel('Substrate Concentration (S)')
ax3.set_ylabel('Reaction Rate')
ax3.set_title('Cooperativity Effect on Rate')
ax3.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax3.grid(True, alpha=0.3)

# 4. Steepness analysis (derivative at S=K_m=1)
ax4 = fig.add_subplot(2, 2, 4)
# Calculate steepness (derivative at S=1)
S_analysis = np.linspace(0.05, 3.5, 200)
steepness_curves = []

colors = ['blue', 'green', 'orange', 'red']

for ind, n in enumerate([0.5, 1.0, 2.0, 4.0]):
    # Numerical derivative
    rate = hill_kinetics(S_analysis, k_max_fixed, K_m, n)
    derivative = np.gradient(rate, S_analysis)
    steepness_curves.append(derivative)
    ax4.plot(S_analysis, derivative, label=f'n={n}', color=colors[ind], linewidth=3)

ax4.axvline(x=1, color='red', linestyle='--', alpha=0.7, label='S=K_m=1')
ax4.set_xlabel('Substrate Concentration (S)')
ax4.set_ylabel('Rate of Change (dRate/dS)')
ax4.set_title('Steepness Analysis\n(Change of Rate)')
ax4.legend()
ax4.grid(True, alpha=0.3)


plt.tight_layout()
plt.show()


# Print analysis
print("Hill Kinetics Transition Analysis")
print("=" * 50)
print(f"Fixed parameters: k_max = {k_max_fixed}, K_m = {K_m}")
print("\nKey observations:")
print("1. n = 1: Classic hyperbolic Michaelis-Menten kinetics")
print("2. n < 1: Negative cooperativity - shallow curves")
print("3. n > 1: Positive cooperativity - sigmoidal curves")
print("4. n >> 1: Switch-like behavior - very steep transition")
print(f"5. Half-maximal rate always occurs at S = K_m = {K_m}")
print("\nThe 3D landscape clearly shows how increasing n transforms")
print("the response from gradual to switch-like behavior!")
```
